```python
import time
import requests
from bs4 import BeautifulSoup
import re
import requests
import time
import numpy as np
import json
import pandas as pd
pd.set_option("max_colwidth", 500)

import unicodedata
import ast
import os
import pickle
from unidecode import unidecode
from pandarallel import pandarallel
```

```python
urls=['https://news.un.org/en/news?page=%s'%i for i in range(6352)]
df = pd.DataFrame(urls,  columns =['url'] )
df
```

```python
def extract_urls(url):
    coverpage = requests.get(url).content
    soup = BeautifulSoup(coverpage, 'html.parser')
    rows = soup.find_all('div', class_= 'view-content')[0].find_all('div', class_ = re.compile('views-row views-row'))
    return [str(i) for i in rows]
```

```python
pandarallel.initialize(nb_workers=32, progress_bar= True)
df[['rows']]  = df['url'].parallel_apply(extract_urls)
df = df.explode('rows').reset_index(drop=True)
df
```

```python
df.to_csv('unnews_all_raw.csv', index=False)
```

```python
def extract_rows(row):
    soup = BeautifulSoup(row, 'html.parser')

    title = soup.find_all('h1', class_ ="story-title")[0].text

    date = soup.find_all('div', class_ ="story-date")[0].text
    date = re.sub("^\s+|\s+$", "", date, flags=re.UNICODE)

    region = soup.find_all('div', class_ ="topics")[0].text
    region = re.sub("^\s+|\s+$", "", region, flags=re.UNICODE)

    abstract = soup.find_all('div', class_ ="news-body")[0].text
    abstract = unicodedata.normalize("NFKD",abstract)
    abstract = re.sub("^\s+|\s+$", "", abstract, flags=re.UNICODE)
    abstract = unicodedata.normalize("NFKD",abstract)

    href = soup.find_all('a', href=True)[0]['href']
    if href.startswith('/'):
        href = 'https://news.un.org'+ href

    return pd.Series([title, date, region, abstract, href])
```

```python
pandarallel.initialize(nb_workers=8)
df[['title', 'date', 'location', 'abstract', 'link']]  = df['rows'].parallel_apply(extract_rows).values
df
```

```python
def extract_html(url):
    time.sleep(2)
    try:
        coverpage = requests.get(url).content
        soup = BeautifulSoup(coverpage, 'html.parser')
    except:
        soup = 'error'
    return str(soup)

def extract_text(html):
    soup = BeautifulSoup(html, 'html.parser')
    find_contents = soup.find_all('div',class_="field-item even")
    if len(find_contents) == 0:
        return 'error'
    else:
        contents = max(find_contents, key=len)

        text = ''
        for paragraph in contents.find_all('p', recursive=False):
            text = text + unicodedata.normalize("NFKD",paragraph.text)
        return text
    
def extract_text(url):
#     time.sleep(1)
    try:
        coverpage = requests.get(url).content
        soup = BeautifulSoup(coverpage, 'html.parser')
    
        find_contents = soup.find_all('div',class_="field-item even")
        if len(find_contents) == 0:
            print('error: %s'%url)
            return 'error'
        else:
            contents = max(find_contents, key=len)
            text = ''
            for paragraph in contents.find_all('p', recursive=False):
                text = text + unicodedata.normalize("NFKD",paragraph.text)
            return text
    except:
        print('error: %s'%url)
        return 'error'
    
def extract_text(url):
#     time.sleep(1)
    try:
        coverpage = requests.get(url).content
        soup = BeautifulSoup(coverpage, 'html.parser')
    
        find_contents= soup.find_all('div',class_="paragraphs-items paragraphs-items-field-news-story paragraphs-items-field-news-story-full paragraphs-items-full")
        if len(find_contents) == 0:
            print('error: %s'%url)
            return 'error'
        else:
            contents = find_contents[0]
            text = ''
            for paragraph in contents.find_all('p', recursive=True):
                text = text + unicodedata.normalize("NFKD",paragraph.text)
            return text
    except:
        print('error: %s'%url)
        return 'error'
```
