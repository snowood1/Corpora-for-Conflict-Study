```python
import requests
from bs4 import BeautifulSoup
import re
import sklearn
import time
import numpy as np
import json
import pandas as pd
pd.set_option("max_colwidth", 500)

# import sys   
# sys.setrecursionlimit(25000)

from pandarallel import pandarallel
```

```python
def extract_row(url):
    try:
        time.sleep(2)
        res=requests.get(url)
        soup = BeautifulSoup(res.content,'html.parser')
    except:
        print('cant connect',url)
        return
    
    rows = soup.find_all('li', {'class': re.compile('views-row views-row.*')})

    if len(rows):
        outputs = []
        for a in rows:
            try:
                href = a.a['href']
                if href.startswith('/'):
                    href = 'https://www.thenewhumanitarian.org' + href

                title = a.a['title']

                date = a.find_all(class_ ='date--short')[0].text
                date = re.sub("^\s+|\s+$", "", date, flags=re.UNICODE)

                topic = a.span.text
                topic = re.sub("^\s+|\s+$", "", topic, flags=re.UNICODE)

            except:
                print(url)

            output = {'title':title, 'date':date,'topic':topic,'href': href}
            outputs.append(output)
    else:
        outputs = None
        
    return outputs

from newspaper import Article
def extract_text(url):
    try:
        time.sleep(2)
        a = Article(url)
        a.download()
        a.parse()
        text = a.text
    except:
        text = ''
    return text
```

```python
prefix =\
'https://www.thenewhumanitarian.org/advanced-search?keys=&&field_report_theme_tid[0]=275' +\
'&field_report_theme_tid[1]=10&field_report_theme_tid[2]=407&field_report_theme_tid[3]=466' +\
'&field_report_theme_tid[4]=447&field_report_theme_tid[5]=463&field_report_theme_tid[6]=5863' +\
'&field_report_type_tid[0]=561&field_report_type_tid[1]=416&field_report_type_tid[2]=15' +\
'&field_report_type_tid[3]=5856&field_report_type_tid[4]=14&field_report_type_tid[5]=6186' +\
'&field_report_type_tid[6]=5359&field_report_type_tid[7]=17&field_authors_tid=' +\
'&date-filter-min-date[date]=&date-filter-max-date[date]=&items_per_page=60&page='


df = pd.DataFrame(range(0, 841,1), columns = ['page'])
df.page = prefix + df.page.astype('str')
df
```

```python
pandarallel.initialize(nb_workers=16)
batch_size = 16*8
i=0
while i<= df.shape[0]:
    print(i, end=',')
    df.loc[i: i+ batch_size-1, 'rows']= df.loc[i: i+ batch_size-1, 'page'].parallel_apply(extract_row).values
    i += batch_size
df.to_csv('newhuman_raw.csv', index=False)
```
