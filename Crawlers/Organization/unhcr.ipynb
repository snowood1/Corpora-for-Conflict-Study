```python
url = 'https://www.unhcr.org/en-us/news-releases.html'
driver = webdriver.Chrome(options=chrome_options, executable_path=chrome_drive)
driver.get(url)
```

```python
all_rows = []
while True:
    try:
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        next_find = driver.find_element_by_xpath("//*[contains(text(), 'Next')]")
        next_find.click()
        time.sleep(2)
        html = driver.page_source
        soup = BeautifulSoup(html, 'html.parser')
        rows = soup.find_all('a', class_= 'cta__box cta__box--s')
        rows = [str(i) for i in rows]
        all_rows = all_rows+ rows
    except:
        break        
```

```python
def extract_rows(row):
    soup = BeautifulSoup(row, 'html.parser')
    
    link = soup.a['href']
    if link.startswith('/'):
        link = 'https://www.unhcr.org/' + link
    
    text =  re.sub("^\s+|\s+$", "", soup.text, flags=re.UNICODE) 
    text = re.sub("\n+", "\n", text, flags=re.UNICODE)
    text = text.split('\n')
    title, date = text[0], text[-1]
    abstract = text[1] if len(text)>2 else ''
    
    return pd.Series([date, title, abstract, link])

#     try:
#         soup = BeautifulSoup(row, 'html.parser')
#         text =  re.sub("^\s+|\s+$", "", soup.text, flags=re.UNICODE) 
#         text = re.sub("\n+", "\n", text, flags=re.UNICODE)
#         text = text.split('\n')
#         title, date = text[0], text[-1]
#         abstract = text[1] if len(text)>3 else ''
#         return pd.Series([date, sec, title, abstract, author, href])
#     except:
#         print('error')
#         return pd.Series(['error']*6)
```

```python
df['text']=None
pandarallel.initialize(nb_workers= 4)
batch_size = 16
i=0
while i< df.shape[0]:
    print(i)
    df.loc[i: i+ batch_size-1, 'text']= df.loc[i: i+ batch_size-1, 'href'].parallel_apply(extract_text).values
    i += batch_size
    if i% (batch_size*5) ==0:
        df.to_csv('unhcr_text.csv',index=False)

df.to_csv('unhcr_text.csv',index=False)
```
