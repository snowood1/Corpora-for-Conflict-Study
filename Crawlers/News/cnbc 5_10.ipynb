```python
years = range(2006, 2013)
months = ['January','February','March','April','May','June','July',\
         'August','September','October','November','December']
days = range(1, 32)

products = itertools.product(years, months, days)

date = ['%s/%s/%s/'%(year, month, day)for (year, month, day) in products]
df= pd.DataFrame(date, columns=['date'])

df['page'] = 'https://www.cnbc.com/site-map/articles/' + df.date
df
```

```python
def get_links(page):
    try:
        coverpage = requests.get(page).content
        soup = BeautifulSoup(coverpage, 'html.parser')
        articles = soup.find_all('a', href=True, class_="SiteMapArticleList-link")
        result = [str(i) for i in articles]
    except:
        result= None
    return result
```

```python
pandarallel.initialize(progress_bar=True)
df[['links']]= df['page'].parallel_apply(get_links)
```

```python
df= df.explode('links').reset_index(drop=True)
df
```

```python
def get_a(html):
    soup = BeautifulSoup(html, 'html.parser')
    a = soup.find_all('a', href=True)
    result = [str(i) for i in a ]
    return result

pandarallel.initialize(progress_bar=False)
df[['a']]= df['links'].parallel_apply(get_a)
```

```python
df= df.explode('a').reset_index(drop=True)
df = df[~df.a.isnull()]
df = df[~df.a.str.contains('class=')]
df = df[~df.a.duplicated()].reset_index(drop=True)
df
```

```python
with open('wiki_relevant.txt', 'r') as file:    
    relevant_rules = file.read().split('\n')
relevant_rules = '\\b' + '|\\b'.join(relevant_rules)

if relevant_rules.endswith('|\\b'):
    relevant_rules =relevant_rules[:-3]   

relevant_rules = re.compile(relevant_rules, re.IGNORECASE)


with open('irelevant_keywords.txt', 'r') as file:    
    irelevant_rules = file.read().split('\n')
irelevant_rules = '\\b' + '|\\b'.join(irelevant_rules)

if irelevant_rules.endswith('|\\b'):
    irelevant_rules =irelevant_rules[:-3]   

irelevant_rules = re.compile(irelevant_rules, re.IGNORECASE)

find_relevant = lambda x: dict(Counter(relevant_rules.findall(str(x).lower())))
find_exclude = lambda x: dict(Counter(irelevant_rules.findall(str(x).lower())))
```

```python
from collections import Counter
df['head_in'] = (df['title'] + ' ' + df['url']).parallel_apply(find_relevant)
df['head_out'] =(df['title'] +' ' + df['url']).parallel_apply(find_exclude)
df['head_in_count']= df['head_in'].str.len()
df['head_out_count']= df['head_out'].str.len()
df
```

```python
head_in = df['head_in_count']
head_out = df['head_out_count']

remove_index =\
    ((head_in ==0) & ( head_out > 0)) |\
    ((head_in > 0) & ( head_out > 1))

remove_index = df[remove_index== True].index
remove_index.shape
```

```python
df = df[~df.index.isin(remove_index)].reset_index(drop=True)
df
```

```python
# import logging
from newspaper import Article, Config


user_agent = 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10.15; rv:78.0) Gecko/20100101 Firefox/78.0'

config = Config()
config.browser_user_agent = user_agent
config.request_timeout = 10

def extract_story(url):
    try:
        article = Article(url,config= config)
        time.sleep(2)
        article.download()
        article.parse()
        return article.text
    except:
        print('error %s'%url)
        return
```
