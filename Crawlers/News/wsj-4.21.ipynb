```python
headers={
    'accept': '*/*',
    'referer': 'https://www.latimes.com/sitemap/',
    'user-agent': 'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/88.0.4324.150 Safari/537.36'
}
   
from newspaper import Article, Config

user_agent = 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10.15; rv:78.0) Gecko/20100101 Firefox/78.0'

config = Config()
config.browser_user_agent = user_agent
config.request_timeout = 10

def extract_story(url):
    try:
        article = Article(url,config= config)
        article.download()
        time.sleep(2)
        article.parse()
        return article.text
    except:
        print('error %s'%url)
        return
    
def extract_text(url):
    try:
        coverpage = requests.get(url, headers = headers).content
        time.sleep(1)
    except:
        print('error url', url)
        return pd.Series[None,None]
    soup = BeautifulSoup(coverpage, 'html.parser')
    bold = soup.find_all('p', class_= 't-content__chapo')
    bold = bold[0].text if len(bold) else ''

    body = soup.find_all('div', class_= 't-content__body u-clearfix')
    if len(body):
        paragraph = body[0].find_all('p')
        paragraph = [p.text for p in paragraph]
        paragraph = '\n'.join(paragraph)
    else:
        paragraph = ''
    text = bold + '\n' + paragraph
    text = re.sub("^\s+|\s+$", "", text, flags=re.UNICODE)
    
    find_tags = soup.find_all('div', class_= 't-content__tags')
    if len(find_tags):
        tags = find_tags[0].find_all('a', href=True)
        tags = '\t'.join([i.text for i in tags])
    else:
        tags=''
    
    if len(text)==0:
        print('error text', url)
        text = 'error text'   
    return pd.Series([text, tags])
```
