# CNN 1998 -2002

```python
urls =\
['https://www.cnn.com/ALLPOLITICS/stories/1998/%02d/'%i for i in range(1,13)]+\
['https://www.cnn.com/ALLPOLITICS/stories/1999/%02d/'%i for i in range(1,13)]+\
['https://www.cnn.com/2000/ALLPOLITICS/stories/%02d/'%i for i in range(1,13)]+\
['https://www.cnn.com/2001/ALLPOLITICS/stories/01/']+\
['https://www.cnn.com/2001/ALLPOLITICS/%02d/'%i for i in range(2,13)]+\
['https://www.cnn.com/2002/ALLPOLITICS/%02d/'%i for i in range(1,13)]

df=pd.DataFrame(urls, columns=['page'])
df
```

```python
page= df.page[0]
coverpage = requests.get(page).content
soup = BeautifulSoup(coverpage, 'html.parser')
ul = soup.find_all('ul')
p = soup.find_all('p')
[str(i) for i in ul+p ]
```

```python
def get_links(page):
    coverpage = requests.get(page).content
    soup = BeautifulSoup(coverpage, 'html.parser')
    ul = soup.find_all('ul')
    p = soup.find_all('p')
    result = [str(i) for i in ul+p ]
    return result
```

```python
pandarallel.initialize(progress_bar=False)
df[['links']]= df['page'].parallel_apply(get_links)
```

```python
df= df.explode('links').reset_index(drop=True)
df
```

```python
def get_a(html):
    soup = BeautifulSoup(html, 'html.parser')
    a = soup.find_all('a', href=True)
    result = [str(i) for i in a ]
    return result

pandarallel.initialize(progress_bar=False)
df[['a']]= df['links'].parallel_apply(get_a)
```

```python
df= df.explode('a').reset_index(drop=True)
df = df[~df.a.isnull()]
df = df[~df.a.str.contains('class=')]
df = df[~df.a.duplicated()].reset_index(drop=True)
df

def extract_a(a):
    a = BeautifulSoup(a, 'html.parser')
    title = re.sub("^\s+|\s+$", "", a.text, flags=re.UNICODE) 
    href= a.a['href']
    if href.startswith('/'):
        href = 'https://www.cnn.com'+ href
    return pd.Series([title, href])

pandarallel.initialize(progress_bar=False)
df[['title','url']]= df['a'].parallel_apply(extract_a)
df = df.iloc[:,[0,3,4]]
df
```

```python
# import logging
from newspaper import Article, Config


user_agent = 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10.15; rv:78.0) Gecko/20100101 Firefox/78.0'

config = Config()
config.browser_user_agent = user_agent
config.request_timeout = 10

def extract_story(url):
    try:
        article = Article(url,config= config)
        article.download()
        article.parse()
        return article.text
    except:
        print('error %s'%url)
        return
```

```python
pandarallel.initialize(nb_workers=128, progress_bar=True)
df['text']= df['url'].parallel_apply(extract_story).values
```

```python
df= df[~df.text.isnull()]
df.text.str.len().describe()
```

```python
df = df[df.text.str.len()>200].reset_index(drop=True)
df
```

# CNN 2011 - 2013

```python
urls =\
['https://www.cnn.com/article/sitemap-2011-%s.html'%i for i in range(1,13)]+\
['https://www.cnn.com/article/sitemap-2012-%s.html'%i for i in range(1,13)]+\
['https://www.cnn.com/article/sitemap-2013-%s.html'%i for i in range(1,4)]
df=pd.DataFrame(urls, columns=['page'])
df
```

```python
def get_links(page):
    coverpage = requests.get(page).content
    soup = BeautifulSoup(coverpage, 'html.parser')
    div = soup.find_all('div', class_='sitemap-entry')[1]
    li = div.find_all('li')
    return [str(i) for i in li]

df['li'] = df['page'].parallel_apply(get_links).values
```

```python
df = df.explode('li').reset_index(drop=True)
```

```python
def extract_li(x):
    soup = BeautifulSoup(x, 'html.parser')
    try:
        date, title = soup.get_text('\n').split('\n')
        href = soup.a['href']
        if href.startswith('/'):
            href = 'https://www.cnn.com' + href
        return pd.Series([date, title, href])
    except:
        return pd.Series([None]*3)

pandarallel.initialize(progress_bar=True)
df[['date', 'title', 'url']] = df['li'].parallel_apply(extract_li).values
```

```python
df =df[~df.url.isnull()].reset_index(drop=True)
df[['text']]=None
# df = df.iloc[:,2:].reset_index(drop=True)
df
```

```python
pandarallel.initialize(nb_workers=64, progress_bar=True)
df['text']= df['url'].parallel_apply(extract_story).values
```

```python
df = df[~df.text.isnull()]
df=df[~df.text.str.startswith('Need a CNN account? ')]
df = df[df.text.str.len()>200]
df
```
